### Ejercicio 1

Para $n=10$ y $n=100$ simular $1000$ veces la variable aleatoria $X_n$ definida como la suma de $n$ variables aleatorias $X_i$ con distribución uniforme en $[0,1]$. Graficar el histograma de las $1000$ realizaciones de $X_n$ y comparar con la densidad de la variable aleatoria $X_n$. ¿Qué observas?

```python
# Para ello debe importar las librerías necesarias
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Genere unos datos cualquiera
x = np.random.normal(0, 1, 1000)

# Genere los datos de la distribución normal con media 0 y desviación estándar 1
xvar = np.random.normal(0, 1, 1000)

# Grafique la distribución de los datos
plt.hist(xvar, bins=20, density=True)
plt.show()
```

```python
# Numero de muestras
n = 1000

# Numero de experimentos
m = 100000

# Esperanza (media)
mu = 0

# Varianza
sigma = 1

# Generar las muestras
x = np.random.normal(mu, sigma, (n, m))

# Calcular la media de las muestras
x_bar = np.mean(x, axis=0)

# Graficar el histograma
plt.hist(x_bar, bins=100, density=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametros
media_exp = 2   # beta= 0.5
n = 10000       # Numero de variables aleatorias
m = 50          # Numero de w's

# Generamos m muestras de n variables aleatorias con distribucion exponencial:
x = np.random.exponential(media_exp, (m,n))
    # (m,n) es una matriz (filas, columnas)

# Obtenemos la media de cada una de las muestras:
x_barra = np.mean(x, axis=1)
    # axis=1: calcula la media de cada fila
    # (m, 1) es un vector

# Obtenemos el segundo momento de cada una de las muestras:
x_barra_cuadrado = np.mean(x_barra**2)

print(x_barra_cuadrado)
```
